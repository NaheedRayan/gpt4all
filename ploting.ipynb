```python


import pandas as pd
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None  # default='warn'

# Step 3: Read CSV File
file_path = 'llmdataset11.csv'
df = pd.read_csv(file_path)



# type(df)
df["T/F"] = ""
df["Res"] = ""

df


x = 0 
y = 0 
for i in range(len(df['Result'])):
    if(df['Result'][i] == True):
        df["T/F"][i] = 1
        x += 1 
    elif(df["Result"][i] == False):
        df["T/F"][i] = 0
        y += 1
    else :
        df["T/F"][i] = 2


df
# # print(df.to_string)
df.plot()
plt.show()

print((x/(x+y))*100)
```
