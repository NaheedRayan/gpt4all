```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Offline', 'Sequential', 'Merged']
values = [60, 77, 59]
colors = ['#FF9999', 'skyblue', 'lightgreen']  # Define colors for each bar

# Create bar chart with custom colors
bars = plt.bar(categories, values, color=colors)

# Set y-axis limit to 100
plt.ylim(0, 100)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Total data')
plt.title('Correct predicted Answers')

# Add values on top of each bar
for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 2, value,
             ha='center', va='bottom')

# Show plot
plt.show()
```
